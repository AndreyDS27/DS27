## Выбросы

Определение целесообразно только для количественных данных, столбцы rectal_temperature и total_protein

```python
q1 = df['rectal_temperature'].quantile(0.25)
q3 = df['rectal_temperature'].quantile(0.75)
irq = q3 - q1
lower_bound = q1 - (1.5*irq)
upper_bound = q3 + (1.5*irq)

remuve_outliers = df[df['rectal_temperature'].between(lower_bound, upper_bound, inclusive=True)]
pd.concat([df, remuve_outliers]).drop_duplicates(keep=False)
```

#### Значения параметра rectal_temperature за пределами границ выброса являются возможными. Поэтому их удаление нецелесообразно.

```python
q1 = df['total_protein'].quantile(0.25)
q3 = df['total_protein'].quantile(0.75)
irq = q3 - q1
lower_bound = q1 - (1.5*irq)
upper_bound = q3 + (1.5*irq)

remuve_outliers = df[df['total_protein'].between(lower_bound, upper_bound, inclusive=True)]
pd.concat([df, remuve_outliers]).drop_duplicates(keep=False)
```

#### Значения параметра total_protein за пределами границ выброса являются возможными. Поэтому их удаление нецелесообразно.
