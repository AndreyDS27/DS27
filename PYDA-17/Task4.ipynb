# Домашнее задание к лекции "Функции"

Вам нужно помочь секретарю автоматизировать работу. Для этого нужно написать программу, которая будет на основе хранимых данных исполнять пользовательские команды.

Исходные данные имеют следующую структуру:

перечень всех документов
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
перечень полок, на которых хранятся документы (если документ есть в documents, то он обязательно должен быть и в directories)
directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}
Общие требования к программе:

код должен быть грамотно декомпозирован (каждая функция отвечает за свою конкретную задачу, дублирующийся функционал переиспользуется, а его код не повторяется);
в коде отсутствуют глобальные переменные (за исключением documents и directories);
пользовательский ввод обрабатывается в цикле while до тех пор, пока пользователь явно не завершит программу (вводом команды "q").

Пример работы:

Результат:

№: 2207 876234, тип: passport, владелец: Василий Гупкин, полка хранения: 1

Примеры работы:

Пункт 5. Пользователь по команде "ds" может удалить существующую полку из данных (только если она пустая)
Примеры работы:

Введите номер полки:

Введите номер полки:

Введите номер полки:

Результат:

```python
# Task 1

documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}


def get_owner_doc(documents, num_doc):
    '''owner_doc(documents, num_doc) - возвращает имя владельца документа.\
    Аргументы функции: documents - список документов, num_doc - номер документа.'''
    d = False
    for doc in documents:
        if doc['number'] == num_doc:
            d = True            
            return print('Результат:\n', doc['name'])            
    if d == False:
        print('Документ не найден в базе')
       
    
def get_shelf(directories, num_doc):
    '''shelf(directories, num_doc) - возвращает полку хранения по номеру документа.\
    Аргументы функции: directories - словарь{полка: номер документа документа}, num_doc - номер документа.'''
    d = False
    for dir in directories:
        if num_doc in directories[dir]:
            d = True
            return print(f'Результат:\n Документ хранится на полке: {dir}')
    if d == False:
        print('Документ не найден в базе')

            
def get_info(documents, directories):
    '''info(documents, directories) - выводит всю информацию по документам'''
    print('Результат:\n')
    for doc in documents:
        number = doc['number']
        for dir in directories:
            if number in directories[dir]:
                print(f"№: {number}, тип: {doc['type']}, владелец: {doc['name']}, полка хранения: {dir}")
    return


def add_shelf(directories, num_shelf):
    '''add_shelf(directories, num_shelf) - добавляет новую полку с номером по запросу.\
    Аргументы функции: directories - словарь{полка: номер документа документа}, shelf_list - номер полки.'''
    if num_shelf in directories:
        shelf_list = list({shelf for shelf in directories})
        print(f'Такая полка уже существует. Текущий перечень полок: {shelf_list}.')
    else:
        directories.setdefault(num_shelf, list())
        shelf_list = list({shelf for shelf in directories})
        print(f'Результат:\n Полка {num_shelf} добавлена. Текущий перечень полок: {shelf_list}')
    return


def del_shelf(directories, num_shelf):
    '''del_shelf(directories, num_shelf) - удаляет полку с номером по запросу.
    Аргументы функции: directories - словарь{полка: номер документа документа}, shelf_list - номер полки.'''
    if num_shelf in directories:
        if len(directories[num_shelf]) == 0:
            directories.pop(num_shelf)
            shelf_list = list({shelf for shelf in directories })
            print(f'Результат:\n Полка {num_shelf} удалена. Текущий перечень полок: {shelf_list}.')
        else:
            shelf_list = list({shelf for shelf in directories})
            print(f'На полке есть документ(ы), удалите их перед удалением полки. Текущий перечень полок: {shelf_list}.')

    else:
        print('Результат: Такой полки не существует.')
    return


def main(documents, directories):
    print('''Список команд:\n 
    "p" - узнать владельца документа по его номеру;\n
    "s" - по номеру документа узнать на какой полке он хранится;\n
    "l" - увидеть полную информацию по всем документам;\n
    "ads" - добавить новую полку;\n
    "ds" - удалить существующую полку из данных (только если она пустая);\n
    "q" - выход.''')
    while True:
        user_input = input('Введите команду: ')
        if user_input == 'p':
            num_doc = input('Введите номер документа: ')
            print(get_owner_doc(documents, num_doc))
        elif user_input == 's':
            num_doc = input('Введите номер документа: ')
            print(get_shelf(directories, num_doc))
        elif user_input == 'l':
            print(get_info(documents, directories))
        elif user_input == 'ads':
            num_shelf = input('Введите номер полки: ')
            print(add_shelf(directories, num_shelf))
        elif user_input == 'ds':
            num_shelf = input('Введите номер полки: ')
            print(del_shelf(directories, num_shelf))
        elif user_input == 'q':
            print('Программа завершена.')
            break

            
main(documents, directories)
```
