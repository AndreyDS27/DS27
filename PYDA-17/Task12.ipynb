## Задание 1

Для датафрейма log из материалов занятия создайте столбец source_type по следующим правилам:

- если источник traffic_source равен yandex или google, то в source_type ставится organic
- для источников paid и email из России - ставим ad
- для источников paid и email не из России - ставим other
- все остальные варианты берем из traffic_source без изменений

```python
import pandas as pd
```

```python
visit_log = pd.read_csv('task12/visit_log.csv', sep=';')
visit_log.head()
```

```python
def source_type(row):
    '''заполнение столбца source_type по правилам:
    - если источник traffic_source равен yandex или google, то в source_type ставится organic
    - для источников paid и email из России - ставим ad
    - для источников paid и email не из России - ставим other
    - все остальные варианты берем из traffic_source без изменений'''
    
    if (row['traffic_source'] == 'yandex') or (row['traffic_source'] == 'google'):
        return 'organic'
    elif (row['traffic_source'] == 'paid') or (row['traffic_source'] == 'email'):
        if row['region'] == 'Russia':
            return 'ad'
        else:
            return 'other'
    else:
        return row['traffic_source']
            
```

```python
# Заполнение столбца source_type в соответствии с функцией source_type
visit_log['source_type'] = visit_log.apply(source_type, axis=1)

visit_log[visit_log['region'] == 'Russia'].head(10)
```

## Задание 3

Используйте файл с оценками фильмов ml-latest-small/ratings.csv. Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок. Под временем жизни понимается разница между максимальным и минимальным значением столбца timestamp для данного значения userId.

```python
ratings = pd.read_csv('ml-latest-small/ratings.csv')
ratings.head()
```

```python
ratings.groupby('userId').count().head()
```

```python
ratings.groupby('userId').agg({'rating': 'count', 'timestamp': ['min', 'max']}).head()
```

```python
results = ratings.groupby('userId').agg({'rating': 'count', 'timestamp': ['min', 'max']})
results['average_lifetime'] = results.apply(lambda row: row['timestamp']['max'] - row['timestamp']['min'], axis=1)
results.head()
```

```python
results[results['rating']['count'] > 100].sort_values(('rating', 'count'), ascending=False).head()
```

Обратите внимание, что в процессе объединения таблиц данные не должны теряться.

Дана статистика услуг перевозок клиентов компании по типам:

rzd - железнодорожные перевозки
auto - автомобильные перевозки
air - воздушные перевозки
client_base - адреса клиентов

```python
auto = pd.DataFrame(
    {
        'client_id': [113, 114, 115, 116, 117],
        'auto_revenue': [57483, 83, 912, 4834, 98]
    }
)
auto
```
