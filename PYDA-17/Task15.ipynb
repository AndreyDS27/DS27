Приведем название столбцов к одному виду и удалим дополнительные столбцы.

```python

df_2019.columns = ["rank","country",
                   "score","gdp_per_capita","social_support",
                   "healthy_life_expectancy","freedom_to_life_choise",
                   "generosity","corruption_perceptions"]
df_2019.head()
```

```python
df_2018.columns = ["rank","country",
                   "score","gdp_per_capita","social_support",
                   "healthy_life_expectancy","freedom_to_life_choise",
                   "generosity","corruption_perceptions"]
df_2018.head()
```

```python
df_2017.drop(['Whisker.high','Whisker.low','Dystopia.Residual'], axis=1, inplace=True)
df_2017.columns = ["country","rank",
                   "score","gdp_per_capita","social_support",
                   "healthy_life_expectancy","freedom_to_life_choise",
                   "generosity","corruption_perceptions"]
df_2017.head()
```

```python
df_2016.drop(['Region','Lower Confidence Interval','Upper Confidence Interval','Dystopia Residual'], axis=1, inplace=True)
df_2016.columns = ["country","rank",
                   "score","gdp_per_capita","social_support",
                   "healthy_life_expectancy","freedom_to_life_choise",
                   "corruption_perceptions","generosity"]
df_2016.head()
```

```python
df_2015.drop(['Standard Error','Region','Dystopia Residual'], axis=1, inplace=True)
df_2015.columns = ["country","rank",
                   "score","gdp_per_capita","social_support",
                   "healthy_life_expectancy","freedom_to_life_choise",
                   "corruption_perceptions","generosity"]
df_2015.head()
```

## Добавим столбец с годом отчета

```python

df_2019['year'] = 2019
df_2018['year'] = 2018
df_2017['year'] = 2017
df_2016['year'] = 2016
df_2015['year'] = 2015
```

## Соедим все датафреймы в один

```python
data = pd.concat([df_2019, df_2018, df_2017, df_2016, df_2015])
data.info()
```

```python
data.head()
```

## Определение пропусков данных

```python
for col in data.columns:
    missing = data[col].isna().mean()
    print(f'{col} - {missing: .1%}')
```

Данные полные и почти не имеют пропусков, за исключением одного значения в столбце 'corruption_perceptions'.

```python
for col in data.columns:
    missing = data[col].isna()
    if any(missing.values):
        display(data[missing])
```

## Заключение

Посмотрим как менялась оценка счастья "score" по годам 

```python
pivot_score = data.pivot_table(
                        index='country', 
                        columns='year', 
                        values='score') 

pivot_score
```

```python
pivot_score.plot(title='Изменение "score" по странам по годам')
```

```python
sns.heatmap(pivot_score)
```

Графики, практически, повторяют форму по годам, что говорит о том, что в большинстве стран никаких значимых изменений за 5 лет не произошло.
