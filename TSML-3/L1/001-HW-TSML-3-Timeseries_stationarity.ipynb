# Домашнее задание 

## Знакомство с временными рядами

Проанализируйте ряды, сделайте их стационарными.

```python
import pandas as pd
import numpy as np

from statsmodels.tsa.stattools import adfuller  # Dickey-Fuller Test - проверка на стационарность временного ряда
import statsmodels.api as sm # для визуализации
import statsmodels.tsa.api as smt # для визуализации

from scipy.stats import boxcox # Box-Cox преобразование

import matplotlib.pyplot as plt
```

## Чтение временных рядов

```python
international_airline_passengers = pd.read_csv('international-airline-passengers.csv')
monthly_car_sales_in_quebec_1960 = pd.read_csv('monthly-car-sales-in-quebec-1960.csv')
monthly_sales_of_company_x_jan_6 = pd.read_csv('monthly-sales-of-company-x-jan-6.csv')
```

```python
all_series = {'international_airline_passengers': international_airline_passengers['International airline passengers: monthly totals in thousands. Jan 49 ? Dec 60'],
              'monthly_car_sales_in_quebec_1960': monthly_car_sales_in_quebec_1960['Count'],
              'monthly_sales_of_company_x_jan_6': monthly_sales_of_company_x_jan_6['Count']}
```

```python
with plt.style.context('bmh'):
    plt.figure(figsize=(16, 8))
    layout = (2, 2)
    for i, key in enumerate(all_series.keys()):
        x = i % 2
        y = int((i - x) / 2)
        
        ts_ax = plt.subplot2grid(layout, (y, x))
        all_series[key].plot(ax=ts_ax, color='blue')
        ts_ax.set_title(key)

    plt.tight_layout()
    
```

## Тест на стационарность

```python
def test_stationarity(timeseries):
    print('Results of Dickey-Fuller Test:')
    dftest = adfuller(timeseries, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic', 'p-value', '#Lags Used', 'Number of Observations Used'])
    for [key, value] in dftest[4].items():
        dfoutput['Critical Value (%s)' % key] = value
    print(dfoutput)
```

```python
def tsplot(y, lags=None, figsize=(14, 8), style='bmh'):
    test_stationarity(y)
    if not isinstance(y, pd.Series):
        y = pd.Series(y)
    with plt.style.context(style):
        plt.figure(figsize=figsize)
        layout = (4, 1)
        ts_ax = plt.subplot2grid(layout, (0, 0), rowspan=2)
        acf_ax = plt.subplot2grid(layout, (2, 0))
        pacf_ax = plt.subplot2grid(layout, (3, 0))

        y.plot(ax=ts_ax, color='blue', label='Or')
        ts_ax.set_title('Original')

        smt.graphics.plot_acf(y, lags=lags, ax=acf_ax, alpha=0.05) # acf - автокорреляция
        smt.graphics.plot_pacf(y, lags=lags, ax=pacf_ax, alpha=0.05) # pcf - частичная автокорреляция

        plt.tight_layout()
    return
```

### Поиск сезонности

```python
def plot_ts_and_points(ts, start_point, step):
    new_series = [None for i in range(len(ts))]
    for i in range(len(ts)):
        pos = start_point + step * i
        if pos >= len(ts):
            break
        new_series[pos] = ts[pos]
    new_series = pd.Series(new_series)
    
    with plt.style.context('bmh'):
        plt.figure(figsize=(16, 8))
        ts_ax = plt.axes()
        ts.plot(ax=ts_ax, color='blue')
        new_series.plot(ax=ts_ax, style='ro')
```

## 1. international_airline_passengers

### Тест Дики-Фуллера и графики коррелограмм

```python
tsplot(all_series['international_airline_passengers'].dropna())
    
```

Тест Дики-Фуллера и графики коррелограмм не отвергают гипотезу о нестационарности ряда (p-value=0.991880, графики автокорреляций выходят за допустимые пределы).

### Поиск сезонности

```python
plot_ts_and_points(all_series['international_airline_passengers'], 2, 4)
```

### Преобразования Бокса-Кокса

```python
series = boxcox(all_series['international_airline_passengers'].dropna(), 0)
```

```python
tsplot(series)
```

удалось убрать размах дисперсии, но тест Дикки-Фуллера все еще не отвергает гипотезу о нестационарности ряда. По графику ряда видно наличие сильного тренда. Уберем его дифференцированием.

### Дифференцирование

```python
series_1 = series[1:] - series[:-1]
```

```python
tsplot(series_1)
```

```python
series_2 = series_1[12:] - series_1[:-12]
```

```python
tsplot(series_2)
```

```python
series_3 = series_2[9:] - series_2[:-9]
```

```python
tsplot(series_3)
```

### Вывод

После преобразований:
p-value = 7.580811e-07, Test Statistic = -5.703497e+00 < Critical Value (1%) = -3.492401e+00.
Соответственно можем отклонить H0 о нестационарности ряда с уровнем значимости менее 1%.
