**1. Найдите спектральное разложение матрицы:**

m = np.array([[1, 2], [2,3]])

```python
m = np.array([[1, 2], [2,3]])
m
```

Спектральное разложение

$$
A = S^T*D*S,
$$
S - ортогональная

D - диагональная из собственных значений.

```python
# w - собственные значения m, v - собственные векторы m
w,v = np.linalg.eig(m)
w,v
```

```python
# S - ортогональная матрица
S = v.T
S
```

```python
S.T
```

```python
# D - диагональная матрица из собственных значений m
D = np.diag(w)
D
```

```python
# проверка
S.T@D@S
```

```python
# проверка
m.tolist() == np.ceil(S.T@D@S).tolist()
```

```python
# проверка
np.allclose(m, S.T@D@S)
```

**2. Постройте зависимость качества восстановления рукописных цифр (MNIST) в зависимости от различного количества компонент PCA. В качестве метрики различия предлагаю использовать MSE.**

У PCA есть метод transform и inverse_transofrm. Первый - снижает размерность, второй - восстанавливает оригинальные данные с потерей информации. Надо посчитать насколько в среднем точно восстанавливается оригинальная картинка в зависимости от количества компонент в PCA.

В качестве справки можно использовать:

https://scikit-learn.org/stable/modules/generated/sklearn.decomposition.PCA.html - Документация по PCA

https://scikit-learn.org/stable/modules/generated/sklearn.metrics.mean_squared_error.html - Документация по MSE

Картинки цифр можно взять отсюда - https://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_digits.html

```python
# Загрузка изображений цифр из sklearn.datasets 

from sklearn.datasets import load_digits
digits = load_digits(n_class = 10 , return_X_y = False , as_frame = False )
print(digits.data.shape)
```

```python
# Импорт Анализ главных компонентов (PCA) из sklearn.decomposition
from sklearn.decomposition import PCA

# импорт Среднеквадратичная ошибка регрессии из sklearn.metrics
from sklearn.metrics import mean_squared_error
```

```python
# Импорт matplotlib.pyplot

import matplotlib.pyplot as plt 

plt.gray() 
plt.matshow(digits.images[0])
```

```python
for i in range(10):
    plt.matshow(digits.images[i])
```

```python
X = digits.data
y = digits.target
```

```python
pca = PCA(n_components=2,
          copy = True,
          whiten = False, 
          svd_solver = 'auto', 
          tol = 0.0, 
          iterated_power = 'auto',
          random_state = 42)
X_reduced = pca.fit_transform(X)
```

```python
# Визуализация для n_components=2

plt.figure(figsize=(10,10))
plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=y, 
            edgecolor='none', alpha=0.7, s=20,
            cmap=plt.cm.get_cmap('nipy_spectral', 10))
plt.colorbar()
plt.title('MNIST. PCA projection')
```

```python
# Расчет точности восстановления (mse - Среднеквадратичная ошибка регрессии) в зависимости от количества компонент в PCA

mse = []

for n in range(digits.data.shape[1]):
    pca = PCA(n_components=n,
          copy = True,
          whiten = False, 
          svd_solver = 'auto', 
          tol = 0.0, 
          iterated_power = 'auto',
          random_state = 42)
    X_reduced = pca.fit_transform(X)              # Уменьшение размерности X
    X_inverse = pca.inverse_transform(X_reduced)  # Преобразование данных обратно в исходное пространство
    mse.append(mean_squared_error(X, X_inverse))  # Среднеквадратичная ошибка регрессии
```

```python
# График зависимости mse от n_components в PCA

plt.figure(figsize=(7,7))
plt.plot(mse, color='b', lw=3, label='mse')
plt.xlabel('Number of components')
plt.ylabel('Mean squared error')
plt.grid()
plt.xlim(0, 63)
plt.title('Зависимость mse от n_components в PCA')
plt.legend()
```
