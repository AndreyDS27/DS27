```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
```

## 1. Нормальная непрерывная случайная величина

```python
# параметры
mu = 2
sigma = 0.1

# Нормальная непрерывная случайная величина
norm_rv = sts.norm(loc=mu, scale=sigma)

# сгенерируем 50 значений
norm_rv.rvs(size=10)
```

Параметр loc задаёт $\mu$, scale — среднеквадратичное отклонение $\sigma$, size — размер выборки. Имя параметра size при вызове функции rvs можно не писать.

Функция `norm_rv.cdf(x)` возвращает значение функции распределения нормальной случайной величины в точке, соответствующей её аргументу:

```python
norm_rv.cdf(3)
```

```python
# график функции распределения
x = np.linspace(-2,4,100)
cdf = norm_rv.cdf(x) # функция может принимать вектор (x)
plt.plot(x, cdf)
plt.ylabel('$F(x)$')
plt.xlabel('$x$')
```

```python
# график плотности распределения
pdf = norm_rv.pdf(x) # функция может принимать и вектор (x)
plt.plot(x, pdf)
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
```

### Графики функции распределения и плотности распределения
### в зависимости от среднеквадратичного отклонения $\sigma$ при $\mu=0$
(Нормальная непрерывная случайная величина)

```python
# Графики функции распределения в зависимости от среднеквадратичного отклонения sigma при mu=0

mu = 0 # сдвигает по оси x
sigma = np.linspace(0.5,5, 5)

for s in sigma:
    norm_rv = sts.norm(loc=mu, scale=s)
    x = np.linspace(-4,4,100)
    cdf = norm_rv.cdf(x)
    plt.plot(x, cdf, label=(f'sigma={s}'))
    plt.ylabel('$F(x)$')
    plt.xlabel('$x$')
    plt.legend()
    
```

```python
# Графики плотности распределения в зависимости от среднеквадратичного отклонения sigma при mu=0

mu = 0
sigma = np.linspace(0.5,5, 5)

for s in sigma:
    norm_rv = sts.norm(loc=mu, scale=s)
    x = np.linspace(-4,4,100)
    pdf = norm_rv.pdf(x)
    plt.plot(x, pdf, label=(f'sigma={s}'))
    plt.ylabel('$f(x)$')
    plt.xlabel('$x$')
    plt.legend()
    
```

## 2. Экспоненциальная непрерывная случайная величина

```python
# параметры
mu = 3
lambda_ = 3

# Экспоненциальная непрерывная случайная величина
exp_rv = sts.expon(loc=mu, scale=lambda_)

# сгенерируем 50 значений
exp_rv.rvs(size=10)
```

Параметр loc задаёт $\mu$, scale — параметр скорости $\lambda$, size — размер выборки. Имя параметра size при вызове функции rvs можно не писать.

Функция `exp_rv.cdf(x)` возвращает значение функции распределения нормальной случайной величины в точке, соответствующей её аргументу:

```python
exp_rv.cdf(3)
```

```python
# график функции распределения
x = np.linspace(-2,4,100)
cdf = exp_rv.cdf(x)
plt.plot(x, cdf)
plt.ylabel('$F(x)$')
plt.xlabel('$x$')
```

```python
# график плотности распределения
pdf = exp_rv.pdf(x) # функция может принимать и вектор (x)
plt.plot(x, pdf)
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
```

### Графики функции распределения и плотности распределения
### в зависимости от параметра $\lambda$ при $\mu=0$
(Экспоненциальная непрерывная случайная величина)

```python
# Графики функции распределения в зависимости от параметра lambda при mu=0

mu = 0 # сдвигает по оси x
lambda_ = np.linspace(1,5, 5)

for lam in lambda_:
    exp_rv = sts.expon(loc=mu, scale=lam)
    x = np.linspace(-4,4,100)
    cdf = exp_rv.cdf(x)
    plt.plot(x, cdf, label=(f'lambda={lam}'))
    plt.ylabel('$F(x)$')
    plt.xlabel('$x$')
    plt.legend()
    
```

```python
# Графики плотности распределения в зависимости от параметра lambda при mu=0

mu = 0 # сдвигает по оси x
lambda_ = np.linspace(1,5, 5)

for lam in lambda_:
    exp_rv = sts.expon(loc=mu, scale=lam)
    x = np.linspace(-4,4,100)
    pdf = exp_rv.pdf(x)
    plt.plot(x, pdf, label=(f'lambda={lam}'))
    plt.ylabel('$f(x)$')
    plt.xlabel('$x$')
    plt.legend()
    
```

## 3. Непрерывная случайная величина Стьюдента

```python
# параметры
n = 2

# Непрерывная случайная величина Стьюдента
t_rv = sts.t(df=n)

# сгенерируем 50 значений
t_rv.rvs(size=10)
```

Параметр df — параметр степеней свободы $n$, size — размер выборки. Имя параметра size при вызове функции rvs можно не писать.

Функция `t_rv.cdf(x)` возвращает значение функции распределения нормальной случайной величины в точке, соответствующей её аргументу:

```python
t_rv.cdf(3)
```

```python
# график функции распределения
x = np.linspace(-4,4,100)
cdf = t_rv.cdf(x)
plt.plot(x, cdf)
plt.ylabel('$F(x)$')
plt.xlabel('$x$')
```

```python
# график плотности распределения
pdf = t_rv.pdf(x) # функция может принимать и вектор (x)
plt.plot(x, pdf)
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
```

### Графики функции распределения и плотности распределения
### в зависимости от параметра степеней свободы n
(Непрерывная случайная величина Стьюдента)

```python
# Графики функции распределения в зависимости от параметра степеней свободы n
N = np.linspace(1,5, 5)
for n in N:
    t_rv = sts.t(df=n)
    x = np.linspace(-4,4,100)
    cdf = t_rv.cdf(x)
    plt.plot(x, cdf, label=n)
    plt.ylabel('$F(x)$')
    plt.xlabel('$x$')
    plt.legend()
    
```

```python
# Графики плотности распределения в зависимости от параметра степеней свободы n
for n in N:
    t_rv = sts.t(df=n)
    x = np.linspace(-4,4,100)
    pdf = t_rv.pdf(x)
    plt.plot(x, pdf, label=n)
    plt.ylabel('$f(x)$')
    plt.xlabel('$x$')
    plt.legend()   
```
