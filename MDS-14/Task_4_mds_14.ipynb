# –î–æ–º–∞—à–Ω–µ–µ –∑–∞–¥–∞–Ω–∏–µ "–ú–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–∏–π –∞–Ω–∞–ª–∏–∑. –ü—Ä–æ–∏–∑–≤–æ–¥–Ω–∞—è"

## –£—Ä–æ–≤–µ–Ω—å 0:¬∂

–ü–æ—Å—á–∏—Ç–∞–π—Ç–µ –∑–Ω–∞—á–µ–Ω–∏–µ –ø–µ—Ä–≤–æ–π –∏ –≤—Ç–æ—Ä–æ–π –ø—Ä–æ–∏–∑–≤–æ–¥–Ω–æ–π —Ñ—É–Ω–∫—Ü–∏–∏:

$f(x) = x^5 + 4sin(2x) + cos(3x+3)$

–í —Ç–æ—á–∫–µ x = 1.

## –£—Ä–æ–≤–µ–Ω—å 1:

–ü–æ—Å—Ç—Ä–æ–π—Ç–µ –≥—Ä–∞—Ñ–∏–∫ –ø–µ—Ä–≤–æ–π –∏ –≤—Ç–æ—Ä–æ–π –ø—Ä–æ–∏–∑–≤–æ–¥–Ω–æ–π —Ñ—É–Ω–∫—Ü–∏–∏:

$f(x) = sin^5(2x+1)$

–ù–∞ –æ—Ç—Ä–µ–∑–∫–µ [-5; 5].

## –£—Ä–æ–≤–µ–Ω—å 0:

–ü–æ—Å—á–∏—Ç–∞–π—Ç–µ –∑–Ω–∞—á–µ–Ω–∏–µ –ø–µ—Ä–≤–æ–π –∏ –≤—Ç–æ—Ä–æ–π –ø—Ä–æ–∏–∑–≤–æ–¥–Ω–æ–π —Ñ—É–Ω–∫—Ü–∏–∏:

$f(x) = x^5 + 4sin(2x) + cos(3x+3)$

–í —Ç–æ—á–∫–µ x = 1.



```python
# –ò–º–ø–æ—Ä—Ç derivative –∏–∑ scipy.misc –¥–ª—è —Ä–∞—Å—á–µ—Ç–∞ –ø—Ä–æ–∏–∑–≤–æ–¥–Ω–æ–π
from scipy.misc import derivative

# –ò–º–ø–æ—Ä—Ç –±–∏–±–ª–∏–æ—Ç–µ–∫–∏ sympy –¥–ª—è –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è —Å–∏–º–≤–æ–ª—å–Ω—ã—Ö –≤—ã—á–∏—Å–ª–µ–Ω–∏–π
from sympy import *
x = symbols('x')
```

#### –ò—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ sympy.diff() –¥–ª—è –Ω–∞—Ö–æ–∂–¥–µ–Ω–∏—è –ø—Ä–æ–∏–∑–≤–æ–¥–Ω–æ–π

```python
# –§–æ—Ä–º—É–ª–∞ 1-–π –ø—Ä–æ–∏–∑–≤–æ–¥–Ω–æ–π c sympy.diff()
dif_1 = diff(ùë•**5+4*ùë†ùëñùëõ(2*ùë•)+ùëêùëúùë†(3*ùë•+3), x)
dif_1
```

```python
# –ó–Ω–∞—á–µ–Ω–∏–µ 1-–π –ø—Ä–æ–∏–∑–≤–æ–¥–Ω–æ–π –≤ x = 1.0
dif_1.evalf(subs={'x':1})
```

```python
# –§–æ—Ä–º—É–ª–∞ 2-–π –ø—Ä–æ–∏–∑–≤–æ–¥–Ω–æ–π c sympy.diff()
dif_2 = diff(dif_1, x)
dif_2
```

```python
# –ó–Ω–∞—á–µ–Ω–∏–µ 2-–π –ø—Ä–æ–∏–∑–≤–æ–¥–Ω–æ–π –≤ x = 1.0
dif_2.evalf(subs={'x':1})
```

#### –ò—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ scipy.misc derivative –¥–ª—è –Ω–∞—Ö–æ–∂–¥–µ–Ω–∏—è –ø—Ä–æ–∏–∑–≤–æ–¥–Ω–æ–π

```python
# –ò—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ scipy.misc –¥–ª—è –Ω–∞—Ö–æ–∂–¥–µ–Ω–∏—è –ø—Ä–æ–∏–∑–≤–æ–¥–Ω–æ–π

'''–û—Ç–¥–∞–µ—Ç —Ñ—É–Ω–∫—Ü–∏—é, –ø—Ä–æ–∏–∑–≤–æ–¥–Ω—É—é –∫–æ—Ç–æ—Ä–æ–π –Ω—É–∂–Ω–æ –Ω–∞–π—Ç–∏'''
def f(x):
    return ùë•**5+4*ùë†ùëñùëõ(2*ùë•)+ùëêùëúùë†(3*ùë•+3) 
```

```python
# –§–æ—Ä–º—É–ª–∞ 1-–π –ø—Ä–æ–∏–∑–≤–æ–¥–Ω–æ–π
derivative(f, x, n=1)
```

```python
# –ó–Ω–∞—á–µ–Ω–∏–µ 1-–π –ø—Ä–æ–∏–∑–≤–æ–¥–Ω–æ–π –≤ x = 1.0
x1 = 1.0
derivative(f, x1, n=1, dx=1e-6)
```

```python
# –§–æ—Ä–º—É–ª–∞ 2-–π –ø—Ä–æ–∏–∑–≤–æ–¥–Ω–æ–π
derivative(f, x, n=2)
```

```python
# –ó–Ω–∞—á–µ–Ω–∏–µ 2-–π –ø—Ä–æ–∏–∑–≤–æ–¥–Ω–æ–π –≤ x = 1.0
x2 = 1.0
derivative(f, x2, n=2, dx=1e-6)
```

***–°–∏–º–≤–æ–ª—å–Ω—ã–µ –≤—ã—Ä–∞–∂–µ–Ω–∏—è –ø—Ä–æ–∏–∑–≤–æ–¥–Ω—ã—Ö –≤ –º–µ—Ç–æ–¥–∞—Ö sympy.diff() –∏ scipy.misc derivative –æ—Ç–ª–∏—á–∞—é—Ç—Å—è, –Ω–æ –≤—ã—á–∏—Å–ª–µ–Ω–∏—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞ –≤ —Ç–æ—á–∫–µ x=1 –∏–¥–µ–Ω—Ç–∏—á–Ω—ã.***

## –£—Ä–æ–≤–µ–Ω—å 1:

–ü–æ—Å—Ç—Ä–æ–π—Ç–µ –≥—Ä–∞—Ñ–∏–∫ –ø–µ—Ä–≤–æ–π –∏ –≤—Ç–æ—Ä–æ–π –ø—Ä–æ–∏–∑–≤–æ–¥–Ω–æ–π —Ñ—É–Ω–∫—Ü–∏–∏:

$f(x) = sin^5(2x+1)$

–ù–∞ –æ—Ç—Ä–µ–∑–∫–µ [-5; 5].

```python
import matplotlib.pyplot as plt
import numpy as np
```

#### –° –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ–º scipy.misc derivative –¥–ª—è –Ω–∞—Ö–æ–∂–¥–µ–Ω–∏—è –ø—Ä–æ–∏–∑–≤–æ–¥–Ω–æ–π

```python
# —Å –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ–º scipy.misc derivative

# –∏–Ω—Ç–µ—Ä–≤–∞–ª x [-5; 5]
x = np.linspace(-5, 5, num=200, endpoint=True)   

'''–û—Ç–¥–∞–µ—Ç —Ñ—É–Ω–∫—Ü–∏—é, –ø—Ä–æ–∏–∑–≤–æ–¥–Ω—É—é –∫–æ—Ç–æ—Ä–æ–π –Ω—É–∂–Ω–æ –Ω–∞–π—Ç–∏'''
def f(x):
    return np.sin(2*x+1)**5

fx = f(x)                                             # –∏—Å—Ö–æ–¥–Ω–∞—è —Ñ—É–Ω–∫—Ü–∏—è
f1x = [derivative(f, xx, dx=1e-6, n=1) for xx in x]   # 1-—è –ø—Ä–æ–∏–∑–≤–æ–¥–Ω–∞—è
f2x = [derivative(f, xx, dx=1e-6, n=2) for xx in x]   # 2-—è –ø—Ä–æ–∏–∑–≤–æ–¥–Ω–∞—è

# –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–æ–≤
plt.figure(figsize=(10, 10))
plt.plot(x, fx, label = 'function')
plt.plot(x, f1x, label = 'f1x')
plt.plot(x, f2x, label = 'f2x')
plt.legend()
plt.title('$f(x) = sin^5(2x+1)$, 1st and 2nd derivative  plots')
```
